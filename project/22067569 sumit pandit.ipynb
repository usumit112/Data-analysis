```python
# Importing necessary libraries
import pandas as pd  # Pandas for data manipulation and analysis
import matplotlib.pyplot as plt # Matplotlib for plotting
# Reading the data from a CSV file named "data.csv" and storing it in a DataFrame called df
df=pd.read_csv("data.csv")
```

```python
df
```

```python
df.drop(columns=['salary','salary_currency'],inplace=True) # Dropping columns 'salary' and 'salary_currency' from the DataFrame df
```

```python
df
```

```python
# Check the NaN counts before dropping
nan_counts_before = df.isna().sum() # Counting NaN values in each column of the DataFrame df insa method in pands helps to detect missing values
print("NaN counts before dropping:")
print(nan_counts_before)

# Drop rows containing NaN values
df_cleaned = df.dropna()

# Check the NaN counts after dropping
nan_counts_after = df_cleaned.isna().sum() # Counting NaN values in each column of the cleaned DataFrame
print("\nNaN counts after dropping:")
print(nan_counts_after)
```

```python
duplicate_row = [df.duplicated()] # Identifying and storing duplicated rows in the DataFrame df into the variable duplicate_row
print (df)
```

```python
for column in df.columns: # For each column in the DataFrame, find and print its unique values
    unique_values = df[column].unique() # Store unique values of the current column
    print(f"Unique values in column '{column}': {unique_values}")
```

```python
# Renaming columns using a dictionary to make them more descriptive
df.rename(columns={
    'SE': 'Senior Level/Expert',
    'MI': 'Medium Level/Intermediate',
    'EN': 'Entry Level',
    'EX': 'Executive Level'
}, inplace=True)

print(df)
```

```python
# Importing skewness and kurtosis functions from scipy.stats
from scipy.stats import skew, kurtosis

# Choose the variable want to analyze
chosen_variable = 'salary_in_usd'

# Calculate summary statistics
summary_stats = df[chosen_variable].describe()
variable_skewness = skew(df[chosen_variable]) # Calculating skewness
variable_kurtosis = kurtosis(df[chosen_variable])  # Calculating kurtosis

# Print summary statistics
print("Summary statistics of {}: \n".format(chosen_variable))
print(summary_stats)
print("\nSkewness: {}".format(variable_skewness))
print("Kurtosis: {}".format(variable_kurtosis))
```

```python


# Calculate correlation between work_year and salary_in_usd
correlation = df['work_year'].corr(df['salary_in_usd'])

# Print correlation
print("Correlation between work_year and salary_in_usd:", correlation)
```

```python


# Get the top 15 jobs
top_jobs = df['job_title'].value_counts().head(15)

# Create a bar graph
plt.figure(figsize=(10, 6)) # Setting figure size
top_jobs.plot(kind='bar') # Plotting a bar graph
plt.title('Top 15 Jobs')  # Setting the title of the plot
plt.xlabel('Job Title') # Setting the label for the x-axis
plt.ylabel('Frequency') # Setting the label for the y-axis
plt.xticks(rotation=45, ha='right') # Rotating x-axis labels for better readability
plt.tight_layout() # Adjusting layout
plt.show() # Displaying the plot
```

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data from a CSV file
df = pd.read_csv('data.csv')

# Group data by job title and calculate the mean salary for each job
average_salary_by_job = df.groupby('job_title')['salary_in_usd'].mean()

# Find the job with the highest average salary
highest_salary_job = average_salary_by_job.idxmax()

# Create a bar graph
plt.figure(figsize=(20, 10)) # Setting figure size
average_salary_by_job.plot(kind='bar') # Plotting a bar graph
plt.title('Average Salary by Job Title') # Setting the title of the plot
plt.xlabel('Job Title')  # Setting the label for the x-axis
plt.ylabel('Average Salary (USD)') # Setting the label for the y-axis
plt.xticks(rotation=55, ha='right') # Rotating x-axis labels for better readability
plt.tight_layout() # Adjusting layout

# Highlight the job with the highest average salary
plt.axhline(y=average_salary_by_job[highest_salary_job], color='r', linestyle='--', label='Highest Salary Job') # Adding a horizontal line
plt.legend() # Adding a legend  It's a way to provide a visual key to help interpret the plot.

plt.show()

print("Job with the highest average salary:", highest_salary_job)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from a CSV file
data = pd.read_csv('data.csv')

# Calculate the average salary for each experience level
average_salary_by_experience = data.groupby('experience_level')['salary_in_usd'].mean().sort_values()

# Create a bar graph
plt.figure(figsize=(10, 6)) # Setting the figure size
average_salary_by_experience.plot(kind='barh', color='skyblue')  # Plotting a horizontal bar graph with sky blue color
plt.title('Average Salary by Experience Level') # Setting the title of the plot
plt.xlabel('Average Salary (USD)') # Setting the label for the x-axis
plt.ylabel('Experience Level') # Setting the label for the y-axis
plt.tight_layout() # Adjusting layout to prevent overlap
plt.show()
```

```python


# Load data from a CSV file
data = pd.read_csv('data.csv')

# Choose the variables you want to analyze
chosen_variables = ['salary_in_usd']

# Create a histogram for each chosen variable
for variable in chosen_variables:
    plt.figure(figsize=(5, 3)) # Setiing the figure size
    plt.hist(data[variable], bins=20, color='skyblue', edgecolor='black')  # Plot histogram with specified bins and colors
    plt.title('Histogram of {}'.format(variable))  # Set title of the histogram
    plt.xlabel(variable) # Set label for x-axis
    plt.ylabel('Frequency') # Set label for y-axis
    plt.grid(True) # Add grid lines
    plt.show()

# Create a box plot for each chosen variable
for variable in chosen_variables:
    plt.figure(figsize=(5, 3))  # Set the figure size
    plt.boxplot(data[variable], vert=False) # Plot boxplot horizontally
    plt.title('Box Plot of {}'.format(variable))  # Set title of the box plot
    plt.xlabel(variable) # Set label for x-axis
    plt.grid(True)  # Add grid line
    plt.show()
```
